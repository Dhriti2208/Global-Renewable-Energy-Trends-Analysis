```python
import numpy as np
import matplotlib.pyplot as plt

# Years in sorted order
years = sorted(grouped_sorted_df['Year'].unique())

# Loop fro each year
for year in years:
    df_year = grouped_sorted_df[grouped_sorted_df['Year'] == year].copy()

    if df_year[['Solar Share (%)', 'Wind Share (%)', 'Hydro Share (%)']].isnull().all().all():
        continue

    # Total share column
    df_year['Total'] = (
        df_year['Solar Share (%)'] +
        df_year['Wind Share (%)'] +
        df_year['Hydro Share (%)']
    )

    country_mix_df_sorted = df_year.sort_values('Total', ascending=False).reset_index(drop=True)

    # Plot in chunks of 50 countries
    chunk_size = 50
    num_chunks = int(np.ceil(len(country_mix_df_sorted) / chunk_size))

    for i in range(num_chunks):
        chunk = country_mix_df_sorted.iloc[i*chunk_size : (i+1)*chunk_size]

        chunk.set_index('Entity')[['Solar Share (%)', 'Wind Share (%)', 'Hydro Share (%)']].plot(
            kind='barh', stacked=True, figsize=(12, 16), colormap='Set2'
        )

        plt.title(f"Renewable Source Mix by Country – Year {year} – Part {i+1}")
        plt.xlabel("Share (%)")
        plt.ylabel("Country")
        plt.xlim(0, 100)
        plt.tight_layout()
        plt.show()
```

```python
# Bar chart of top 10 per-capita leaders in renewable generation, with before/after comparison (e.g., 2010 vs. 2023).
df_bar = grouped_sorted_df.copy()

base_year = 2010 if 2010 in df_bar['Year'].unique() else df_bar['Year'].min()
latest_year = 2023 if 2023 in df_bar['Year'].unique() else df_bar['Year'].max()

df_base = df_bar[df_bar['Year'] == base_year]
df_latest = df_bar[df_bar['Year'] == latest_year]

top10_latest = df_latest.nlargest(10, 'Renewable Capacity per Capita (MW)')[['Entity', 'Renewable Capacity per Capita (MW)']]
top10_countries = top10_latest['Entity'].tolist()

comparison = pd.DataFrame({'Entity': top10_countries})
comparison = comparison.merge(df_base[df_base['Entity'].isin(top10_countries)][['Entity', 'Renewable Capacity per Capita (MW)']],
                              on='Entity', how='left').rename(columns={'Renewable Capacity per Capita (MW)': str(base_year)})

comparison = comparison.merge(df_latest[df_latest['Entity'].isin(top10_countries)][['Entity', 'Renewable Capacity per Capita (MW)']],
                              on='Entity', how='left').rename(columns={'Renewable Capacity per Capita (MW)': str(latest_year)})

plot_df = comparison.melt(id_vars='Entity', value_vars=[str(base_year), str(latest_year)],
                          var_name='Year', value_name='Capita (MW)')

plt.figure(figsize=(12,6))
sns.barplot(data=plot_df, x='Entity', y='Capita (MW)', hue='Year')
plt.title(f"Top 10 Per‑Capita Renewable Capacity: {base_year} vs {latest_year}")
plt.xlabel("Country")
plt.ylabel("Renewable Capacity per Capita (MW)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
